```python
!pip install scikit-fuzzy
# Librerias
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
# Definición de las variables de entrada y salida
tamaño = ctrl.Antecedent(np.arange(300,601,0.01),'tamaño')
defectos = ctrl.Antecedent(np.arange(0,20,0.01),'defectos')
sabor = ctrl.Antecedent(np.arange(0,1,0.001),'sabor')
calidad = ctrl.Consequent(np.arange(0,5,0.01),'calidad')

# Definición de las funciones de pertenencia de entrada
tamaño['4'] = fuzz.trapmf(tamaño.universe,[300,315,335,350])
tamaño['3'] = fuzz.trapmf(tamaño.universe,[340,370,420,450])
tamaño['2'] = fuzz.trapmf(tamaño.universe,[420,470,500,550])
tamaño['1'] = fuzz.trapmf(tamaño.universe,[500,520,600,600])
tamaño.view()

defectos['a-'] = fuzz.trapmf(defectos.universe,[0,0.5,1,1.5])
defectos['a'] = fuzz.trapmf(defectos.universe,[1,3,4,5])
defectos['a+'] = fuzz.trapmf(defectos.universe,[4.5,10,20,20])
defectos.view()

sabor['dulce'] = fuzz.trapmf(sabor.universe,[0,0,0.5,1])
sabor['acida'] = fuzz.trapmf(sabor.universe,[0.75,1,1,1])
sabor.view()

# Definición de la función de pertenencia de salida
calidad['IV'] = fuzz.trapmf(calidad.universe,[3,3.5,4,4])
calidad['III'] = fuzz.trapmf(calidad.universe,[1.5,2,2.5,3.2])
calidad['II'] = fuzz.trapmf(calidad.universe,[0.5,1,1.5,2])
calidad['I'] = fuzz.trapmf(calidad.universe,[0,0,0.5,1])
calidad.view()

# Reglas difusas
rule1=ctrl.Rule(tamaño['4']|defectos['a-']&sabor['dulce'],calidad['IV'])
rule2=ctrl.Rule(tamaño['4']&defectos['a-']&sabor['acida'],calidad['IV'])
rule3=ctrl.Rule(tamaño['4']&defectos['a']&sabor['dulce'],calidad['IV'])
rule4=ctrl.Rule(tamaño['4']&defectos['a']&sabor['acida'],calidad['IV'])
rule5=ctrl.Rule(tamaño['4']&defectos['a+']&sabor['dulce'],calidad['IV'])
rule6=ctrl.Rule(tamaño['4']&defectos['a+']&sabor['acida'],calidad['IV'])
rule7=ctrl.Rule(tamaño['3']&defectos['a-']&sabor['dulce'],calidad['III'])
rule8=ctrl.Rule(tamaño['3']&defectos['a-']&sabor['acida'],calidad['IV'])
rule9=ctrl.Rule(tamaño['3']&defectos['a']&sabor['dulce'],calidad['IV'])
rule10=ctrl.Rule(tamaño['3']&defectos['a']&sabor['acida'],calidad['IV'])
rule11=ctrl.Rule(tamaño['3']&defectos['a+']&sabor['dulce'],calidad['IV'])
rule12=ctrl.Rule(tamaño['3']&defectos['a+']&sabor['acida'],calidad['IV'])
rule13=ctrl.Rule(tamaño['2']&defectos['a-']&sabor['dulce'],calidad['II'])
rule14=ctrl.Rule(tamaño['2']&defectos['a-']&sabor['acida'],calidad['II'])
rule15=ctrl.Rule(tamaño['2']&defectos['a']&sabor['dulce'],calidad['III'])
rule16=ctrl.Rule(tamaño['2']&defectos['a']&sabor['acida'],calidad['IV'])
rule17=ctrl.Rule(tamaño['2']&defectos['a+']&sabor['dulce'],calidad['IV'])
rule18=ctrl.Rule(tamaño['2']&defectos['a+']&sabor['acida'],calidad['IV'])
rule19=ctrl.Rule(tamaño['1']&defectos['a-']&sabor['dulce'],calidad['I'])
rule20=ctrl.Rule(tamaño['1']&defectos['a-']&sabor['acida'],calidad['I'])
rule21=ctrl.Rule(tamaño['1']&defectos['a']&sabor['dulce'],calidad['I'])
rule22=ctrl.Rule(tamaño['1']&defectos['a']&sabor['acida'],calidad['II'])
rule23=ctrl.Rule(tamaño['1']&defectos['a+']&sabor['dulce'],calidad['II'])
rule24=ctrl.Rule(tamaño['1']&defectos['a+']&sabor['acida'],calidad['III'])

# Resultados del Sistema Difuso
type_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,	rule13,	rule14,	rule15,	rule16,	rule17,	rule18,	rule19,	rule20,	rule21,	rule22,	rule23,	rule24])
type = ctrl.ControlSystemSimulation(type_ctrl)
type.input['tamaño']=int(input('Peso de tu fruta (300-600g):'))
type.input['defectos']=float(input('Área de defectos de tu fruta (0-20cm2):'))
type.input['sabor']=float(input('Sabor de tu fruta (0-1):'))
type.compute()
print('La calidad de tu fruta es (entre 1 y 4): ',str(int(type.output['calidad'])))
calidad.view(sim=type)
```

```python
# Lo hice de esta manera en la práctica anterior por lo que voy a hacer copia y pega 

import numpy as np
import skfuzzy as fuzz
import skfuzzy.control as ctrl

# Definición de las variables
temperatura = ctrl.Antecedent(np.arange(0, 41, 1), 'temperatura')
humedad = ctrl.Antecedent(np.arange(0, 101, 1), 'humedad')
termostato = ctrl.Consequent(np.arange(-2, 3, 1), 'termostato')

# Funciones de pertenencia para temperatura
temperatura['frio'] = fuzz.trimf(temperatura.universe, [0, 0, 15])
temperatura['normal'] = fuzz.trimf(temperatura.universe, [10, 20, 30])
temperatura['calor'] = fuzz.trimf(temperatura.universe, [25, 40, 40])
temperatura.view()


# Funciones de pertenencia para humedad
humedad['baja'] = fuzz.trimf(humedad.universe, [0, 0, 33])
humedad['media'] = fuzz.trimf(humedad.universe, [0, 50, 100])
humedad['alta'] = fuzz.trimf(humedad.universe, [67, 100, 100])

humedad.view()

# Funciones de pertenencia para termostato
termostato['baja'] = fuzz.trimf(termostato.universe, [-2, -1, 0])
termostato['mantiene'] = fuzz.trimf(termostato.universe, [-1, 0, 1])
termostato['sube'] = fuzz.trimf(termostato.universe, [0, 1, 2])


termostato.view()

# Definición de las reglas
regla1 = ctrl.Rule(temperatura['frio'], termostato['sube'])
regla2 = ctrl.Rule(temperatura['normal'] & ~humedad['alta'], termostato['mantiene'])
regla3 = ctrl.Rule(temperatura['calor'] | humedad['alta'], termostato['baja'])

# Crear el sistema de control
sistema_control = ctrl.ControlSystem([regla1, regla2, regla3])

# Crear una simulación del sistema de control
simulacion = ctrl.ControlSystemSimulation(sistema_control)

# Probar la simulación con diferentes entradas
simulacion.input['temperatura'] = 18
simulacion.input['humedad'] = 40

# Calcular la salida
simulacion.compute()

print(simulacion.output['termostato'])
```

Clases:

```python
import numpy as np
import skfuzzy as fuzz
import skfuzzy.control as ctrl

# Antecedentes
sepal_length = ctrl.Antecedent(np.arange(4, 8, 0.01), 'sepal_length')
sepal_width = ctrl.Antecedent(np.arange(2, 5, 0.01), 'sepal_width')
petal_length = ctrl.Antecedent(np.arange(0, 7, 0.01), 'petal_length')
petal_width = ctrl.Antecedent(np.arange(0, 3, 0.01), 'petal_width')

# Consecuente
class_ = ctrl.Consequent(np.arange(0, 4, 1), 'class')

# Funciones de pertenencia
sepal_length.automf(3)  # Bajo, Medio, Alto
sepal_width.automf(3)   # Bajo, Medio, Alto
petal_length.automf(3)  # Bajo, Medio, Alto
petal_width.automf(3)   # Bajo, Medio, Alto

class_['setosa'] = fuzz.trimf(class_.universe, [0, 1, 2])
class_['versicolor'] = fuzz.trimf(class_.universe, [1, 2, 3])
class_['virginica'] = fuzz.trimf(class_.universe, [2, 3, 4])
```

```python
rule1 = ctrl.Rule(sepal_length['poor'] & petal_length['poor'], class_['setosa'])
rule2 = ctrl.Rule(sepal_length['average'] & petal_length['average'], class_['versicolor'])
rule3 = ctrl.Rule(sepal_length['good'] & petal_length['good'], class_['virginica'])
```

### Voy a usar el algoritmos genéticos

```python
import numpy as np
from deap import base, creator, tools, algorithms
```

```python
import pandas as pd

# Cargue el dataset en un DataFrame de pandas
df = pd.read_csv('Iris.csv')
data = df.values.tolist()
features = df.drop('Species', axis=1).values.tolist()
labels = df['Species'].tolist()
```

```python
def configure_membership_functions(individual):
    """Configura las funciones de pertenencia basadas en los parámetros del individuo."""
    mf_params = {}
    idx = 0
    for feature in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
        mf_params[feature] = {
            'setosa': individual[idx:idx+3],
            'versicolor': individual[idx+3:idx+6],
            'virginica': individual[idx+6:idx+9]
        }
        idx += 9
    return mf_params

def classify(sample, mf_params):
    """Clasifica una muestra basada en las funciones de pertenencia."""
    scores = {}
    for species in ['setosa', 'versicolor', 'virginica']:
        score = 1.0
        for i, feature in enumerate(['sepal_length', 'sepal_width', 'petal_length', 'petal_width']):
            # Añadir declaraciones de impresión para depurar
            print(f"Evaluating feature: {feature}, species: {species}")
            print(f"Sample value: {sample[i]}")
            print(f"Membership function parameters: {mf_params[feature][species]}")
            
            score *= interp_membership([0, 5, 10], mf_params[feature][species], sample[i])
        scores[species] = score
    return max(scores, key=scores.get)


def evaluate(individual):
    """Evalúa el individuo."""
    mf_params = configure_membership_functions(individual)
    
    correct = 0
    for sample in data:
        prediction = classify(sample[:-1], mf_params)
        if prediction == sample[-1]:
            correct += 1

    accuracy = correct / len(data)
    return accuracy,

def interp_membership(x, xmf, value):
    """
    Interpolated membership function lookup.

    Parameters:
        x (array-like): Array of domain values.
        xmf (array-like): Array of membership function values. Must be the same length as `x`.
        value (float): Value to get the interpolated membership of.

    Returns:
        float: Interpolated membership value.
    """
    if value <= x[0]:
        return xmf[0]
    if value >= x[-1]:
        return xmf[-1]

    for i, xi in enumerate(x):
        if xi == value:
            return xmf[i]
        if xi > value:
            if i == 0:
                continue
            m = (xmf[i] - xmf[i - 1]) / (x[i] - x[i - 1])
            b = xmf[i] - m * x[i]
            return m * value + b
```

```python
# Define el problema como un problema de maximización
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

toolbox = base.Toolbox()
toolbox.register("attr_float", np.random.uniform, 0, 1)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_float, n=36) 
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

```python
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.1)
toolbox.register("select", tools.selTournament, tournsize=3)
toolbox.register("evaluate", evaluate)

population = toolbox.population(n=50)
NGEN=40
for gen in range(NGEN):
    offspring = algorithms.varAnd(population, toolbox, cxpb=0.5, mutpb=0.1)
    fits = toolbox.map(toolbox.evaluate, offspring)
    for fit, ind in zip(fits, offspring):
        ind.fitness.values = fit
    population = toolbox.select(offspring, k=len(population))
```

```python
# 1. Selección del Mejor Individuo
best_ind = tools.selBest(population, 1)[0]
print("Mejor individuo: ", best_ind)
mf_params_best = configure_membership_functions(best_ind)

# 2. Realizar Predicciones
def predict(samples, mf_params):
    predictions = []
    for sample in samples:
        predictions.append(classify(sample, mf_params))
    return predictions

# Datos para hacer nuevas predicciones
new_samples = [
    [5.1, 3.5, 1.4, 0.2], # Esperado: setosa
    [6.7, 3.0, 5.2, 2.3], # Esperado: virginica

]

predicted_labels = predict(new_samples, mf_params_best)
print(predicted_labels)
```
