**Exercise 2:** Create a program named 'grayscale.py' that reads a color image and saves it in grayscale. The program will receive the color image file name and the grayscale image file name as arguments. The syntax of 'argparse' for this exercise is provided:


If the image cannot be loaded or saved, the program should print the message "Error loading the image" or "Error saving the image," respectively.

```python
import cv2
import argparse

def grayscale(color_image_file, grayscale_image_file):
    # Read the color image
    color_image = cv2.imread(color_image_file)

    # Check if the image was successfully loaded
    if color_image is None:
        print("Error loading the image")
        return

    # Convert the color image to grayscale
    grayscale_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

    # Save the grayscale image
    success = cv2.imwrite(grayscale_image_file, grayscale_image)

    # Check if the image was successfully saved
    if not success:
        print("Error saving the image")
        return

    print("Grayscale image saved successfully")

# Example usage
color_image_file = "lena.jpg"
grayscale_image_file = "output.jpg"
grayscale(color_image_file, grayscale_image_file)
```

```python
import cv2
#from google.colab import drive

# Mount Google Drive
#drive.mount('/content/gdrive')

# Create a VideoCapture object for the webcam
cap = cv2.VideoCapture(0)

# Get the width and height of the frames
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_path = 'output.avi'
out = cv2.VideoWriter(output_path, fourcc, 20.0, (width, height))

while True:
    # Capture frame by frame
    ret, frame = cap.read()

    # Process the frames here
    if ret:
        # Example processing: convert to grayscale
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Write the frame to the output video file
        out.write(gray_frame)
    else:
        break

    # Break the loop when the user presses 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the VideoWriter and VideoCapture objects
out.release()
cap.release()
```

**Brightness and Contrast**

**Exercise 1:** Implement an OpenCV program called `bc.py` that receives the name of an image as a parameter (to be loaded in grayscale) and displays its brightness and contrast.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def bc(image_path):
    # Load the image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if img is None:
        print("Error loading the image")
        return

    # Calculate the mean and standard deviation of the image
    mean = np.mean(img)
    std = np.std(img)

    # Calculate the brightness and contrast
    brightness = mean
    contrast = std

    # Display the brightness and contrast
    print("Brightness:", brightness)
    print("Contrast:", contrast)

    # Show the image
    plt.imshow(img, cmap='gray')
    plt.title('Grayscale Image')
    plt.axis('off')
    plt.show()

# Example usage
image_path = "lena.jpg"
bc(image_path)
```

**Exercise 2:** Create a program named 'histogramColor.py' that, given a three-channel image, displays (and saves to an image) histograms for each of its basic colors instead of in grayscale.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def histogramColor(image_path):
    # Read the image
    image = cv2.imread(image_path)

    if image is None:
        print("Error reading the image")
        return

    # Split the image into its color channels
    b, g, r = cv2.split(image)

    # Calculate the histograms for each color channel
    hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
    hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
    hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])

    # Display the histograms
    plt.figure(figsize=(12, 6))

    plt.subplot(131), plt.plot(hist_b, color='blue')
    plt.title('Blue Histogram'), plt.xlabel('Pixel Value'), plt.ylabel('Frequency')

    plt.subplot(132), plt.plot(hist_g, color='green')
    plt.title('Green Histogram'), plt.xlabel('Pixel Value'), plt.ylabel('Frequency')

    plt.subplot(133), plt.plot(hist_r, color='red')
    plt.title('Red Histogram'), plt.xlabel('Pixel Value'), plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

    # Save the histograms to an image
    hist_image = np.zeros((256, 768, 3), dtype=np.uint8)
    hist_image[:, :256, 0] = hist_b
    hist_image[:, 256:512, 1] = hist_g
    hist_image[:, 512:, 2] = hist_r

    cv2.imwrite('histogram_image.jpg', hist_image)

# Example usage
image_path = "lena.jpg"
histogramColor(image_path)
```

**Exercise 3:** Create a program named 'colorLight.py' that takes the name of an image as a parameter and extracts specific channels, saving them with specified filenames:


*   L channel of  CIELab, file cielab_l.jpg.
*   V channel of HSV, file hsv_v.jpg.
*   L channel of HSL, file hsl_l.jpg.
*   Y channel of YCbCr, file ycrcb_y.jpg.






```python
import cv2

def colorLight(image_path):
    # Read the image
    image = cv2.imread(image_path)

    if image is None:
        print("Error reading the image")
        return

    # Convert the image to CIELab color space
    cielab_image = cv2.cvtColor(image, cv2.COLOR_BGR2Lab)

    # Extract the L channel of CIELab
    cielab_l = cielab_image[:, :, 0]

    # Save the L channel of CIELab
    cv2.imwrite("cielab_l.jpg", cielab_l)

    # Convert the image to HSV color space
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Extract the V channel of HSV
    hsv_v = hsv_image[:, :, 2]

    # Save the V channel of HSV
    cv2.imwrite("hsv_v.jpg", hsv_v)

    # Convert the image to HSL color space
    hsl_image = cv2.cvtColor(image, cv2.COLOR_BGR2HLS)

    # Extract the L channel of HSL
    hsl_l = hsl_image[:, :, 1]

    # Save the L channel of HSL
    cv2.imwrite("hsl_l.jpg", hsl_l)

    # Convert the image to YCbCr color space
    ycrcb_image = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)

    # Extract the Y channel of YCbCr
    ycrcb_y = ycrcb_image[:, :, 0]

    # Save the Y channel of YCbCr
    cv2.imwrite("ycrcb_y.jpg", ycrcb_y)

# Example usage
image_path = "lena.jpg"
colorLight(image_path)
```

**Exercise 4**: Write a complete program containing this example code to visualize the pseudocolored result of an image.

```python
import cv2
import matplotlib.pyplot as plt

def pseudocolor_image(image_path):
    # Read the image
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply a colormap (pseudocoloring)
    colored_image = cv2.applyColorMap(gray_image, cv2.COLORMAP_HOT)

    # Display the original and pseudocolored images
    plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
    plt.subplot(122), plt.imshow(cv2.cvtColor(colored_image, cv2.COLOR_BGR2RGB)), plt.title('Pseudocolored')
    plt.show()

# Example usage
image_path = 'lena.jpg'
pseudocolor_image(image_path)
```
