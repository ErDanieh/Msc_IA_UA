# Ejemplo de cálculo ID3

- ID3 selecciona la mejor característica en cada paso mientras construye un árbol de decisión.
- Para medir el desorden se utiliza la Entropía $$Entropy(S) = - \sum_{i \in N} p_i * log_2(p_i)$$ siendo $p_i$ la probabilidad.
- En nuestro caso la entropía es 0 si todos los valores de la columna objetivo son homogéneos (similares) y 1 si ocurre todo los contrario.
- Para conocer la ganancia Ganancia de Información de cada característica (atributo) a cada paso $$IG(S, A) = Entropy(S) - \sum ( \frac{|S_v|}{|S|} \times Entropy(S_v))$$ donde $S_v$ las filas de $S$ para cada columna $A$ que tiene una valor $v$, $|S_v|$ o $|S|$ es el número de filas.

## Paso 1

- En el conjunto inicial, $S$, tenemos 14 filas.
- En la variable objetivo, **Infected**, tenemos 8 valores de **Yes** y 6 del **No**.
- Con lo que la entropía inicial de $S$ es: $$Entropy(S) = - (8/14) * log_2(8/14) - (6/14) * log_2(6/14) = 0.99$$
- La ganacia:
$$IG(S, A) = Entropy(S) - \sum ( \frac{|S_v|}{|S|} \times Entropy(S_v))$$ donde $S_v$ son los elementos de $S$ que contienen los valores $v$ de la columna $A$; y $|S_v|$ o $|S|$ respresentan el tamaño de cada conjunto.


```python
def Entropy(S, debug=False): # Columna de un DataFrame que es tipo Series
  '''
  Calcula la entropía para una serie de datos (columna de un DataFrame)
  '''

  res = S.value_counts(normalize=True).apply(lambda x : - x * np.log2(x)).sum()
  if debug:
    print('Entropy')
    print(S)
    df = pd.DataFrame()
    df['n'] = S.value_counts()
    df['x'] = S.value_counts(normalize=True)
    df['log2(x)'] = df['x'].apply(lambda x : np.log2(x))
    df['-x log2(x)'] = -df['x'] * df['log2(x)']
    print(df)
    print(res)
  return res

def IG(S,A,debug=False): # Ganancia (S:Series, A:Series)
  '''
  Calcula la ganacia de cada variable de entrada (A) respecto de la objetivo (S)
  '''

  df = pd.DataFrame({S.name:S,A.name:A})
  if debug:
    print('IG')
    print(df)

  return Entropy(S,debug) + df.groupby(A.name).apply(lambda x:  -len(x) / len(S) * Entropy(x[S.name], debug=debug)).sum()
```

```python
# Example of COVID-19 set
data = pd.read_csv(path_covid19, index_col=0)

# Compute variable Fever (A) with respect Infected (S)
target = 'Infected' # Infected, Play
S = data[target]
A = data['Fever'] # Fever, Weather
IG(S,A)
```

## Paso 2

### Ganancia de **Fever*

- En columna **Fever** tenemos 8 filas con **Yes** y 6 con **No**.
- De las cuales la variable objetivo **Infected**:
   - `Fever = Yes`,  6 de `Infected = Yes` y 2 de `Infected = No`
   - `Fever = No`,  2 de `Infected = Yes` y 4 de `Infected = No`

**Cálculo de entropías:**

$$|S| = 14$$

$$Para \quad v = Si, |S_v| = 8$$

$$Entropy(S_{Si}) = - (6/8) * log_2(6/8) - (2/8) * log_2(2/8) = 0.81$$

$$Para \quad v = NO, |S_v| = 6$$

$$Entropy(S_{No}) = - (2/6) * log_2(2/6) - (4/6) * log_2(4/6) = 0.91$$

**Cálculo de la ganancia:**

$$IG(S, Fever) = Entropy(S) - (|S_{Si}| / |S|) * Entropy(S_{Si}) - (|S_{No}| / |S|) * Entropy(S_{No})$$

$$IG(S, Fever) = 0.99 - (8/14) * 0.81 - (6/14) * 0.91 = 0.13$$


```python
IG(data['Infected'],data['Fever'])
```

```python
def ID3(data, x_names, y_name, depth=0, debug=False):
    tabs = '\t' * depth
    
    # Paso 1: Si todos los ejemplos son de la misma clase, retornar esa clasificación
    if len(data[y_name].unique()) == 1:
        print(f"{tabs}=> {data[y_name].iloc[0]}")
        return

    # Paso 2: Si no hay más características, retornar la clase mayoritaria
    if not x_names:
        print(f"{tabs}=> {data[y_name].mode()[0]}")
        return
    
    # Paso 3: Seleccionar la característica con la mayor ganancia de información
    gains = {x: IG(data[y_name], data[x]) for x in x_names}
    best_feature = max(gains, key=gains.get)
    
    if debug:
        print(f"{tabs}Ganancias de información: {gains}")
        print(f"{tabs}Mejor característica: {best_feature}")
    
    print(f"{tabs}{best_feature} chosen for next node")

    # Paso 4: Crear los nodos hijos y aplicar ID3 recursivamente
    for value in data[best_feature].unique():
        print(f"{tabs}{best_feature} = {value}")
        
        # Crear un subconjunto de datos donde best_feature tiene el valor actual
        subset = data[data[best_feature] == value].drop(columns=[best_feature])
        
        # Aplicar ID3 al subconjunto con las características restantes
        ID3(subset, [x for x in x_names if x != best_feature], y_name, depth + 1, debug)

# Ejecutar el algoritmo ID3 con el conjunto de datos y las características disponibles
ID3(data, x_names, y_name, debug=False)
```
