  **b.5)** Dado un grupo de cereales, se desea estudiar además qué relaciones hay entre los diferentes productos en la categoría. Por ejemplo, se desea conocer dentro del clúster cuáles son los productos más parecidos entre sí, y cuál difiere más del resto de los elementos en grupo. Pudiera proponer cómo obtener esta información utilizando algoritmos de agrupamiento.

```python
from sklearn.cluster import KMeans

# Calculate the mean values for each attribute
prototype = df.mean()

print(prototype)
```

```python

from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances
from scipy.cluster.hierarchy import dendrogram, linkage

# Aplica el algoritmo K-means para agrupar los cereales
kmeans = KMeans(n_clusters=5)
kmeans.fit(df[['calorias', 'proteinas', 'grasas', 'sodio', 'fibra', 'carbohidratos', 'potasio']])

# Obtiene los centroides de cada grupo
centroids = kmeans.cluster_centers_

# Calcula la distancia entre cada par de cereales dentro de cada grupo
distances = cdist(df[['calorias', 'proteinas', 'grasas', 'sodio', 'fibra', 'carbohidratos', 'potasio']], centroids, 'euclidean')

# Encuentra los productos más similares entre sí dentro de cada grupo
most_similar = []
for i in range(5):
    group_distances = distances[:, i]
    most_similar_indices = group_distances.argsort()[:2]  # Obtiene los índices de los dos cereales más similares
    most_similar.append((most_similar_indices[0], most_similar_indices[1]))

# Encuentra el cereal que difiere más del resto de los elementos en cada grupo
most_dissimilar = []
for i in range(5):
    group_distances = distances[:, i]
    most_dissimilar_index = group_distances.argmax()  # Obtiene el índice del cereal más disimilar
    most_dissimilar.append(most_dissimilar_index)

# Imprime los resultados
for i in range(5):
    print(f"Grupo {i+1}:")
    print(f"Productos más similares: {df.iloc[most_similar[i][0]]['nombre']}, {df.iloc[most_similar[i][1]]['nombre']}")
    print(f"Producto más disimilar: {df.iloc[most_dissimilar[i]]['nombre']}")
    print()
    



# Calculate the pairwise distances
distances = pairwise_distances(centroids)

# Plot the pairwise distances
plt.imshow(distances, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Pairwise Distances')
plt.show()



# Calcula la matriz de enlace utilizando el método de enlace completo
Z = linkage(df[['calorias', 'proteinas', 'grasas', 'sodio', 'fibra', 'carbohidratos', 'potasio']], method='complete')

# Plotea el dendrograma
plt.figure(figsize=(10, 6))
dendrogram(Z)
plt.title('Dendrograma de los grupos de cereales')
plt.xlabel('Índices de los cereales')
plt.ylabel('Distancia')
plt.show()
```
